### 4-2. 식품섭취조사 (p133)
~ kstrata : 기본 변수 / N_DIET 식사요법여부 ~

개인당 1개 이상의 행을 가지고 있으므로 유의하여 분석해야 한다. 
- 행(row) : 개인별, 끼니별, 음식별, 식품재료별로 별도의 행을 이루도록 구성되어 있음

```python
## 영양조사
pd_sas = pd.read_sas('data/hn20_24rc.sas7bdat', format='sas7bdat')
pd_sas.shape 
```

```python
pd_sas.head(5)
```

```python
# 식품섭취조사 해당 변수만 뽑기 (기본변수 제외 작업)
ones = list(pd_sas.columns[24:])
ones.append('ID')
#ones
```

```python
# ID 출력 (only)
count_ID = pd_sas.groupby('ID').size()
count_ID.shape # 5805개의 ID 가지고 있음
```

```python
pd_sas.groupby('year').size() # 2020년만 가지고 있는 것 확인!
```

기본DB + 영양조사DB (merge)

```python
merge_test = pd.merge(age_65,pd_sas.loc[:,ones], on = 'ID')
merge_test.head()
```

```python
merge_test.shape # 중복 행때문에 뻥튀기되어서 merge됨
```

이 부분에 대해서는 영양조사 데이터를 어떻게 활용 할지에 대해 정해 볼 필요가 있다

```python
s = merge_test.groupby('ID').size()
s.shape # ID를 1385개 보유하고 있음
```

우선 2020년 데이터 기반 우리가 사용할 수 있는 1차적 데이터의 개수는 1385개이다.
